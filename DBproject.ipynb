# join table

```python
cur.execute("""SELECT arabic.ayah_id,
ayah.arabic,
en_surah.name,
en_surah.translation,
ayah.tot_ayah,
ayah.types,
arabic.text,
en_read.transliteration,
en_trans.text,
indo_trans.translation
FROM arabic
JOIN en_surah ON arabic.surah = en_surah.surah_id 
JOIN ayah ON arabic.surah = ayah.surah
JOIN en_read ON arabic.verse_number = en_read.verse_number
JOIN en_trans ON arabic.verse_number = en_trans.index
JOIN indo_trans ON arabic.verse_number = indo_trans.verse_num
LIMIT 10""")

rows = cur.fetchall()

df = pd.DataFrame(rows, columns=["ayah","arabic_name","eng_name","surah_trans","total_ayah","types","arabic","read","en_trans","indo_trans"])

df
```

# the Question

## How many ayah in quran?

```python
# Execute the query
cur.execute("SELECT SUM(tot_ayah) AS total_ayah FROM ayah")

result = cur.fetchone()

# Print the total number of ayah
print("Total number of ayah in the Quran:", result[0])
```

## How many surah that have ayah less than 5 ayah? 

```python
conn.commit()

cur.execute("""
    SELECT en_surah.name
    FROM (
        SELECT surah, COUNT(*) as tot_ayah
        FROM ayah
        GROUP BY surah
        HAVING tot_ayah < 5
    ) as subquery
    JOIN en_surah ON subquery.surah = en_surah.surah_id;
""")

results = cur.fetchall()

if results:
    print("Surah(s) with less than 5 ayah:")
    for row in results:
        print(row[0])
else:
    print("No surah with less than 5 ayah found.")
```

## which ayah there is the shortest ayah? 

```python
cur.execute("""SELECT ayah_id, text, LENGTH(text) as length
               FROM arabic
               ORDER BY length ASC
               LIMIT 1;
            """)

result = cur.fetchone()
print("The shortest ayah is:", result[1])
```

```python
cur.execute("""SELECT arabic.ayah_id, arabic.text, LENGTH(arabic.text) AS length, arabic.surah
               FROM arabic
               ORDER BY length ASC
               LIMIT 1;
            """)

result = cur.fetchone()
print(f"The shortest ayah is: {result[1]} in Surah {result[3]} Ayah {result[0]}")
```

## which ayah there is the longest ayah?

```python
cur.execute("""
SELECT arabic.ayah_id, arabic.text, LENGTH(arabic.text) AS length, en_surah.name, arabic.surah
FROM arabic
JOIN en_surah ON arabic.surah = en_surah.surah_id
ORDER BY length DESC
LIMIT 1;
""")

result = cur.fetchone()
print("The longest ayah is: {} in Surah {} Ayah {}".format(result[1], result[3], result[0]))
```
